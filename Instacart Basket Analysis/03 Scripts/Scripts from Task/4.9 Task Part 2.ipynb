### Line Chart

```python
# Use sampling to create a line chart
```

```python
# Create a list holding True/False values to the test np.random.rand() <= 0.7

dev = np.random.rand(len(ords_prods)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big

big = ords_prods[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = ords_prods[~dev]
```

```python
# Check that big + small = ords_prods_flag

len(ords_prods)
```

```python
len(big) + len(small)
```

```python
# Reduce dataframe to only the needed columns

df_2 = small[['orders_day_of_week', 'prices']]
```

```python
# Create a line chart

line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
plt.title('Revenue by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Revenue')
```

### 3. Create a histogram of the "order_hour_of_day" column.

```python
# Create a histogram for the amount of orders by time of day

hist_order_daytime = ords_prods['order_hour_of_day'].plot.hist(bins = 24)
plt.title('Number of Orders throughout the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
```

The busiest hours for orders are from morning (~ 9) until late afternoon (~ 17) with a steady decline until the evening. From midnight to about 5 am is the time with the least orders.

### 4. Create a bar chart from the "loyalty_flag" column.

```python
# Create a bar chart for the different loyalty categories

bar2 = ords_prods['loyalty_flag'].value_counts().plot.bar()
plt.title('Distribution of Customers in Different Loyalty Categories')
plt.xlabel('')
plt.ylabel('Number of Customers')
```

Most customers fall into the category "Regular customer", while the category "New customer" has the least amount of people.

### 5. Check whether there's a difference in expenditure (the 'prices' column) depending on the hour of the day.

```python
# Reduce dataframe to only the needed columns

df_3 = small[['order_hour_of_day', 'prices']]
```

```python
# Create a line chart

line_prices_orders_daytime = sns.lineplot(data = df_3, x = 'order_hour_of_day', y = 'prices')
plt.title('Revenue by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Revenue')
```

As you can see, the expenditure fluctuates strongly during the day with the most money being spend around 3 or 4 in the morning and another peak around 10 pm. The least money is being spend at 10 am.

### 6. Determine whether there's a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents.

```python
# Reduce dataframe to only the needed columns

df_4 = small[['age', 'n_dependants']]
```

```python
# Create a line chart

line_age_dependents = sns.lineplot(data = df_4, x = 'age', y = 'n_dependants')
plt.title('Number of Dependants in Relation to Age')
plt.xlabel('Age')
plt.ylabel('Number of Dependants')
```

From this line chart, you can see that there is no obvious relationship between the age of a person and the number of dependents.

### 7. To visualize the relationship between age and spending power (income), create a scatterplot.

```python
# Reduce dataframe to only the needed columns

df_5 = small[['age', 'income']]
```

```python
# Create a scatterplot

scatter_age_income = sns.scatterplot(x = 'age', y = 'income', data = df_5)
plt.title('Income in Relation to Age')
plt.xlabel('Age')
plt.ylabel('Income')
```

Again, you can see, that there is no obvious relationship between age and income.

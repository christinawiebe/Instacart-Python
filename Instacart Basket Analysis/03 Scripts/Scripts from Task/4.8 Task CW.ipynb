### 2. Find the aggregated mean of the "order_number" column grouped by "department_id" for the entire dataframe.

```python
# Group by "department_id" and calculate the mean for "order_number"

ords_prods.groupby('department_id').agg({'order_number' : ['mean']})
```

### 4. Create a loyalty flag for existing customers using transform () and loc().

```python
# Group ords_prods by "user_id" and create a column with the max order numbers

ords_prods['max_order'] = ords_prods.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Look at top rows of the new column

ords_prods.head(15)
```

```python
# Create a flag that assigns a "loyalty label" to a user ID

ords_prods.loc[ords_prods['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods.loc[(ords_prods['max_order'] <= 40) & (ords_prods['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods.loc[ords_prods['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check the loyalty frequency

ords_prods['loyalty_flag'].value_counts(dropna = False)
```

### 5. Do the prices of products purchased by loyal customers differ from those purchased by regular or new customers?

```python
# Group by loyalty categories, then get the mean, min and max
# from "prices"

ords_prods.groupby('loyalty_flag').agg({'prices' : ['mean']})
```

##### The differences in spending between the three different groups aren't too big. But it's interesting to note that new customers seem to be spending the most on average.

### 6. Create a spending flag for each user based on the average price across all their orders.

```python
# Look at top rows of the new column

ords_prods.head(10)
```

```python
# Group ords_prods by "user_id" and create a column with the avg spending

ords_prods['avg_spending'] = ords_prods.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Create a flag that assigns a "spending label" to a user ID

ords_prods.loc[ords_prods['avg_spending'] < 10, 'spending_flag'] = 'Low spender'
ords_prods.loc[ords_prods['avg_spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check the spender frequency

ords_prods['spending_flag'].value_counts(dropna = False)
```

### 7. Create an order frequency flag that marks the regularity of a user's ordering behaviour.

```python
# Group ords_prods by "user_id" and create a column with the median of days since prior order

ords_prods['median_days_since_prior_order'] = ords_prods.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Look at the top rows of the new column

ords_prods.head()
```

```python
# Create a flag that assigns a "frequency label" to a user ID

ords_prods.loc[ords_prods['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods.loc[(ords_prods['median_days_since_prior_order'] > 10) & (ords_prods['median_days_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods.loc[ords_prods['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check the frequency of order frequency

ords_prods['frequency_flag'].value_counts(dropna = False)
```

##### There are 5 rows that couldn't be assigned a flag.

```python
# Find the rows with the NaN in the "frequency_flag" column

ords_prods[ords_prods['median_days_since_prior_order'].isna()]
```

##### This is one customer who ordered 5 different products in one order. Therefore, the only value for "days_since_prior_order" there is for this customer is "NaN". And since there can't be a median of NaN, this customer didn't get a frequency flag.

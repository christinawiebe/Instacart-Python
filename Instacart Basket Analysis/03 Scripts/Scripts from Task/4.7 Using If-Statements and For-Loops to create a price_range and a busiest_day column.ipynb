# If-Statements

```python
# Define the function

def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Use function on "ords_prods_merge" to create a price_range column

subset['price_range'] = subset.apply(price_label, axis = 1)
```

```python
# Check values in new column

subset['price_range'].value_counts()
```

```python
# Look for max price

subset['prices'].max()
```

# If-Statement with loc()

```python
# Create conditions

subset.loc[subset['prices'] > 15, 'price_range_loc'] = 'High range product'
subset.loc[(subset['prices'] <= 15) & (subset['prices'] > 5), 'price_range_loc'] = 'Mid range product'
subset.loc[subset['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
# Check the price range frequency

subset['price_range_loc'].value_counts(dropna = False)
```

```python
# Apply the loc() function on the entire dataframe

ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High range product'
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (subset['prices'] > 5), 'price_range_loc'] = 'Mid range product'
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
# Check the price range frequency

ords_prods_merge['price_range_loc'].value_counts(dropna = False)
```

# For-loop

```python
# Check order number by day

ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create a for-loop for busyness of day

result = []

for value in ords_prods_merge["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
# Print result list

result
```

```python
# Combine result list with ords_prods_merge

ords_prods_merge['busiest_day'] = result
```

```python
# Check the frequency of busyness

ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

### 2. Create a new column for "Busiest days" with the two busiest days of the week and the two slowest days.

```python
# Create a for-loop for busyness of day with adjusted conditions:
# Two busiest and two slowest days

result1 = []

for value in ords_prods_merge["orders_day_of_week"]:
    if value in [0, 1]:
        result1.append("Busiest day")
    elif value in [3,4]:
        result1.append("Least busy")
    else:
        result1.append("Regularly busy")
```

```python
# Print result1

result1
```

```python
# Combine result1 with ords_prod_merge

ords_prods_merge['busiest_days'] = result1
```

```python
# Check the frequency of busyness

ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

##### The "Busiest day" now includes two days and the number corresponds to the numbers of the two busiest days added up. The same is true for the least busiest days. And the regularly busy days are still the most.

### 4. Create a new column "busiest_period_of_day".

```python
# Check order number by hour

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

##### Most Orders: 10h - 16h
##### Fewest Orders: 2h - 5h

```python
# Create a for-loop for business by hour

result2 = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 12, 13, 14, 15, 16]:
        result2.append("Most orders")
    elif value in [2, 3, 4, 5]:
        result2.append("Fewest orders")
    else:
        result2.append("Average orders")
```

```python
# Print result2

result2
```

```python
# Combine result2 with ords_prods_merge

ords_prods_merge['busiest_period_of_day'] = result2
```

```python
# Check the frequency of busyness

ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

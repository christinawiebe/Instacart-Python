### 3. Is there a difference in spending habits between the different U.S. regions?

```python
# Create a "region" column that divides the "state" column into geographical regions


result = []

for value in ic ['state']:
    if value in ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']:
        result.append('northeast')
    elif value in ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']:
        result.append('midwest')
    elif value in ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']:
        result.append('south')
    elif value in ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']:
        result.append('west')
    else:
        result.append('unknown')
```

```python
# Print result list

result
```

```python
# Combine result list with "ic"

ic['region'] = result
```

```python
# Check the frequency of regions

ic['region'].value_counts(dropna = False)
```

```python
# Create a crosstable between "region" and "spending_flag"

region_spending = pd.crosstab(ic['region'], ic['spending_flag'], dropna = False)
```

```python
region_spending
```

All the regions have more low spenders than they have high spenders. The region with the most high spenders is the south, the region with the least high spenders is the northeast. This is most likely due to the size of those regions which results in different sizes in population and therefore possible customers.
